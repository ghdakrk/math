```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f2(x, y):
    return (1-x)**2 + 100 *(y - x**2)**2

xx = np.linspace(-4, 4, 800)
yy = np.linspace(-3, 3, 600)
X, Y = np.meshgrid(xx, yy)
Z = f2(X, Y)

levels = np.logspace(-1, 3, 10)
plt.contourf(X, Y, Z, alpha=0.2, levels=levels)
plt.contour(X, Y, Z, colors='gray', levels=[0.4,3,15,50,150,500,1500,5000])
plt.plot(1, 1 ,'ro', markersize=10)

plt.xlim(-4, 4)
plt.ylim(-3, 3)
plt.xticks(np.linspace(-4, 4, 9))
plt.yticks(np.linspace(-3, 3, 7))
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('2차원 로젠브록 함수 $f(x,y)$')
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
import time
def f1(x):
    return (x-2)**2 + 2
def f_prime(x):
    return 2*(x-2)
x0 = 0
a = time.time()
result = minimize(f1, x0, jac=lambda x: (x-2))  # jac를 지정하면 7배 정도 빠름을 알 수 있다.
b = time.time()
print(result, (b-a)*10000)
```

```python
import scipy as sp

def f2(x):
    return (1-x[0])**2 + 100*(x[1]-x[0]**2)**2

def f2g(x):
    return np.array((2*(x[0]-1) - 400 * x[0] * (x[1]-x[0]**2), 200*(x[1] - x[0]**2)))

x0 = (2, 0.3)
result = sp.optimize.minimize(f2, x0, jac=f2g)

print(result)
```

```python
import scipy.optimize as optimize

def f(x):
      return x[0]**2 + x[1]**2

x0 = np.array([1,1])

def g(x):
      return x[0] + x[1] - 1

optimize.fmin_slsqp(f, x0, eqcons=[g])
```

```python
import scipy.optimize as optimize

def f(x):
      return (x[0]-4)**2 + (x[1]-2)**2

x0 = np.array([1,1])

k = 1
def g1(x):
      return np.atleast_1d(k - np.sum(np.abs(x)))

optimize.fmin_slsqp(f, x0, ieqcons=[g1])
```

```python
import cvxpy as cp
```

```python
a = cp.Variable()
b = cp.Variable()

constraints = [
    a >= 100,
    b >= 100,
    a + 2 * b <=500, 
    4 * a + 5 * b <= 9800, 
]

obj = cp.Maximize(3*a + 5*b)
prob = cp.Problem(obj, constraints)

prob.solve(), prob.status, a.value, b.value
```
