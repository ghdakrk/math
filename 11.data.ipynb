```python
import scipy as sp

def f(x):
    if x<0:
        return 0 * x
    elif 0<=x<180:
        return 1/270 * x
    elif 180<=x<360:
        return 1/540 * x
    else:
        return 0

sp.integrate.quad(f,0,360)
```

```python
total = np.random.randint(1,100,10000)  # 모집단의 총 데이터 수는 1000개
```

```python
total.mean() # 모집단의 평균
```

```python
sample_mean_ls = []  # 표본평균 모음

for i in range(50):  # 표본 수집 50번 반복
    sampling = np.random.randint(0,len(total),100) # 50 개의 표본데이터 추출
    sample = []  # 샘플데이터
    for num in sampling:
        sample.append(total[num])

    sample_mean_ls.append(np.mean(sample))
```

```python
displot(sample_mean_ls)
```

```python
np.mean(sample_mean_ls)
```

```python
# 여기에 코드를 작성해주세요.
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 

x = np.linspace(-100, 500, 1000)
p = np.zeros_like(x)
#### 1. 코드를 작성해주세요 ####
p[(0 < x) & (x <= 180)] = 2 / (3*180)
p[(180 < x) & (x <= 360)] = 1 / (3*180)
xp = x * p

plt.subplot(121)
#### 2. 코드를 작성해주세요 ####
plt.plot(x, p)
plt.xticks([0, 180, 360])
plt.title('$p(x)$')
plt.xlabel('$x$ (도)')

plt.subplot(122)
#### 3. 코드를 작성해주세요 ####
plt.plot(x, xp)
plt.xticks([0, 180, 360])
plt.title('$xp(x)$')
plt.xlabel('$x$ (도)')

plt.show()
```
