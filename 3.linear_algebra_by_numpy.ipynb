```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

plt.rc("font", family="Malgun Gothic")
mpl.rcParams['axes.unicode_minus'] = False

iris = load_iris()
n = 10
x = iris.data[:n, :2]

plt.plot(x.T, 'o:')
plt.xticks(range(2), ['꽃받침 길이', '꽃받침 폭'])
plt.xlim(-0.5, 2)
plt.ylim(2.5, 6)
plt.title('붓꽃 크기 특성')
plt.legend(['표본 {}'.format(i+1) for i in range(n)])
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd

plt.figure(figsize=(8,8))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(x), s=100, color='.2', marker='s')
for i in range(n):
    ax.text(x[i, 0]- 0.05, x[i, 1] + 0.03, f"표본 {i+1}")

plt.xlabel("꽃받침 길이")
plt.ylabel("꽃받침 폭")
plt.title('붓꽃 크기 특성 (2차원)')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA

pca1 = PCA(n_components=1)
x_low = pca1.fit_transform(x)
x2 = pca1.inverse_transform(x_low)

plt.figure(figsize=(7,7))
ax = sns.scatterplot(0, 1, data=pd.DataFrame(x), s=100, color='.2', marker='s')

for i in range(n):
    d = 0.03 if x[i, 1] > x2[i, 1] else -0.04
    ax.text(x[i,0] - 0.065, x[i,1]+d, f"표본{i+1}")
    plt.plot([x[i,0], x2[i,0]], [x[i,1], x2[i,1]], 'k--')

plt.show()
```

```python
from sklearn.datasets import fetch_olivetti_faces
faces_all = fetch_olivetti_faces()
k = 9
faces = faces_all.images[faces_all.target == k]

N = 2
M = 5
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(faces[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle('올리베티 얼굴 사진')
plt.tight_layout()
plt.show()
```

```python
faces.shape, 64*64  # 10개 이미지, 64*64
```

```python
from sklearn.decomposition import PCA
pca3 = PCA(n_components=3)  # 차원 감소
X3 = faces_all.data[faces_all.target == 9]
W3 = pca3.fit_transform(X3)
X32 = pca3.inverse_transform(W3)
```

```python
X32.shape, W3.shape  # 4096 > 2 > 4096
```

```python
N = 2
M = 5
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(X32[k].reshape(64,64), cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
plt.suptitle('주성분 분석으로 그사화한 올레베티 얼굴 사진')
plt.tight_layout()
plt.show()
```

```python
face_mean = pca3.mean_.reshape(64,64)
face_p1 = pca3.components_[0].reshape(64, 64)
face_p2 = pca3.components_[1].reshape(64, 64)
face_p3 = pca3.components_[2].reshape(64, 64)
plt.figure(figsize=(10,10))

plt.subplot(141)
plt.imshow(face_mean, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title('평균 얼굴')

plt.subplot(142)
plt.imshow(face_p1, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title('주성분 1')

plt.subplot(143)
plt.imshow(face_p2, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title('주성분 2')

plt.subplot(144)
plt.imshow(face_p3, cmap=plt.cm.bone)
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title('주성분 3')

plt.show()
```

```python
N = 2
M = 5
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i * M + j
        ax = fig.add_subplot(N, M, k+1)
        w = 1.5 * (k-5) if k < 5 else 1.5 * (k-4)
        ax.imshow(face_mean + w * face_p1, cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(f'주성분 1의 비중{w}')
plt.suptitle('평균 얼굴에 주성분 1을 더한 사진')
plt.tight_layout()
plt.show()
```
